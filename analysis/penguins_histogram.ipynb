# Penguins histogram
This notebook loads the penguins dataset and plots a histogram of bill length.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

data_path = Path("../data/penguins.csv")
df = pd.read_csv(data_path)
bill_lengths = df["bill_length_mm"].dropna()

plt.figure(figsize=(7, 4))
bill_lengths.hist(bins=15, color="steelblue", edgecolor="black")
plt.title("Penguins bill length distribution")
plt.xlabel("Bill length (mm)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
